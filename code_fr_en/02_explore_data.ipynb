# WMT14 FR-EN Dataset Exploration
Load and explore the 2000-example validation subset

```python
# Import all required libraries
from datasets import load_from_disk
import pandas as pd
import random
import matplotlib.pyplot as plt
import pickle
```

## 1. Load the saved dataset

```python
# Load the saved dataset
dataset = load_from_disk("../data/wmt14_fr_en_validation_2000")
print(f"Loaded {len(dataset)} sentence pairs")
print(f"\nFeatures: {dataset.features}")
print(f"Column names: {dataset.column_names}")
```

## 2. Convert to convenient formats

```python
# Convert to list of dictionaries (easiest for iteration)
data_list = []
for item in dataset:
    data_list.append({
        'english': item['translation']['en'],
        'french': item['translation']['fr']
    })

print(f"Created list with {len(data_list)} sentence pairs")
```

```python
# Convert to pandas DataFrame (good for analysis)
df = pd.DataFrame(data_list)
print(f"DataFrame shape: {df.shape}")
df.head()
```

## 3. Explore the data

```python
# Show random examples
print("=" * 80)
print("Random Examples:")
print("=" * 80)

for i in random.sample(range(len(data_list)), 5):
    print(f"\nExample {i}:")
    print(f"FR: {data_list[i]['french']}")
    print(f"EN: {data_list[i]['english']}")
```

```python
# Compute sentence lengths
df['en_length'] = df['english'].str.split().str.len()
df['fr_length'] = df['french'].str.split().str.len()

print("Sentence Length Statistics:")
print("\nEnglish:")
print(df['en_length'].describe())
print("\nFrench:")
print(df['fr_length'].describe())
```

```python
# Visualize length distributions
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df['en_length'], bins=30, alpha=0.7, color='blue', edgecolor='black')
axes[0].set_xlabel('Sentence Length (words)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('English Sentence Lengths')
axes[0].grid(alpha=0.3)

axes[1].hist(df['fr_length'], bins=30, alpha=0.7, color='red', edgecolor='black')
axes[1].set_xlabel('Sentence Length (words)')
axes[1].set_ylabel('Frequency')
axes[1].set_title('French Sentence Lengths')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Compare EN vs FR lengths
plt.figure(figsize=(8, 6))
plt.scatter(df['en_length'], df['fr_length'], alpha=0.5)
plt.xlabel('English Length (words)')
plt.ylabel('French Length (words)')
plt.title('English vs French Sentence Lengths')
plt.plot([0, df['en_length'].max()], [0, df['en_length'].max()], 'r--', alpha=0.5, label='y=x')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

correlation = df['en_length'].corr(df['fr_length'])
print(f"\nCorrelation between EN and FR lengths: {correlation:.3f}")
```

## 4. Export convenient formats for later use

```python
# Save as pickle for quick loading
with open('../data/sentence_pairs_fr_en.pkl', 'wb') as f:
    pickle.dump(data_list, f)

print("âœ“ Saved sentence pairs as pickle: ../data/sentence_pairs_fr_en.pkl")
print("\nTo load later: data_list = pickle.load(open('../data/sentence_pairs_fr_en.pkl', 'rb'))")
```

- `df`: Pandas DataFrame with columns ['english', 'french', 'en_length', 'fr_length']
